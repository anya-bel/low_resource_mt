```python
with open('datasets/evenki/src-train14r.txt') as src, open('datasets/evenki/tgt-train14r.txt') as tgt:
    src = [x.strip() for x in src.readlines()]
    tgt = [x.strip() for x in tgt.readlines()]
```

```python
evenki = pd.DataFrame({'src':src, 'tgt':tgt})
```

```python
unique_russian_words = list(set(sum([x.split() for x in evenki.tgt.to_list()], [])))
```

```python
masculine_words = []
feminin_words = []
for word in unique_russian_words:
    parsed = morph.parse(word)[0]
    postag = parsed.tag.POS
    animacy = parsed.tag.animacy

    if postag == 'NOUN' and animacy == 'anim':
        gender = parsed.tag.gender
        if gender == 'femn':
            feminin_words.append(parsed.normal_form)
        elif gender == 'masc':
            masculine_words.append(parsed.normal_form)
    
```

```python
print(sorted(list(set(feminin_words))))
```
