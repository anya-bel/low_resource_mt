```python
with open('datasets/evenki/src-train14r.txt') as src, open('datasets/evenki/tgt-train14r.txt') as tgt:
    src = [x.strip() for x in src.readlines()]
    tgt = [x.strip() for x in tgt.readlines()]
```

```python
evenki = pd.DataFrame({'src':src, 'tgt':tgt})
```

```python
unique_russian_words = list(set(sum([x.split() for x in evenki.tgt.to_list()], [])))
```

```python
from pymorphy2 import MorphAnalyzer
```

```python
morph = MorphAnalyzer()
```

```python
masculine_words = []
feminin_words = []
for word in unique_russian_words:
    parsed = morph.parse(word)[0]
    postag = parsed.tag.POS
    animacy = parsed.tag.animacy

    if postag == 'NOUN' and animacy == 'anim':
        gender = parsed.tag.gender
        if gender == 'femn':
            feminin_words.append(parsed.normal_form)
        elif gender == 'masc':
            masculine_words.append(parsed.normal_form)
    
```

```python
print(sorted(list(set(feminin_words))))
```

```python
repl_src = ['этырко̄н', 'бэе', 'пастух', 'доктор', 'дядя', 'охотник', 'ама̄ка̄', 'амӣ']
repl_tgt = ['старик', 'мужчина', 'пастух', 'доктор', 'дядя', 'охотник', 'дед', 'отец']
repl_list = list(zip(repl_src, repl_tgt))
```

```python

testdf = pd.DataFrame({'src':[],'new_src':[],'tgt':[] }) 
with open('datasets/evenki_augmented.txt', 'a') as augmented_df:
    for num, row in evenki[(evenki.src.str.contains('нуӈан ')) & (evenki.tgt.str.contains('она '))].iterrows():
        src = row.src
        word_src, word_tgt = random.choice(repl_list)
        new_src = row.src.replace('нуӈан ', f'{word_src} ')
        tgt = row.tgt.replace('он ', f'{word_tgt} ')
        testdf.loc[num] = src, new_src, tgt
        #augmented_df.write(f"{src}, {new_src}, {tgt}\n")
        
testdf
```

```python
for num, row in testdf.iterrows():
    with open('datasets/evenki_augmented.txt', 'a') as augmented_df:
        augmented_df.write(f"{row.src}, {row.new_src}, {row.tgt}\n")


```
