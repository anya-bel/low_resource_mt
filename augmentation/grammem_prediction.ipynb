```python
import requests
from bs4 import BeautifulSoup
```

```python
txt = requests.get('http://siberian-lang.srcc.msu.ru/ru/text/kak-moya-mama-lechila-o-n-borisova').content
```

```python
main_link = 'http://siberian-lang.srcc.msu.ru'
txt2 = requests.get('http://siberian-lang.srcc.msu.ru/ru/textspage').content
soup2 = BeautifulSoup(txt2)
links = set([x.get('href') for x in soup2.findAll('div', class_='view-content')[0].findAll('a')])
```

```python
soup = BeautifulSoup(txt)
```

```python
lex_corpus = []
gram_corpus = []
```

```python
for link in links:
    txt3 = requests.get(main_link+link).content
    soup3 = BeautifulSoup(txt3)
    for sent in soup3.findAll('div', class_='group-gloss-wrap field-group-html-element')[0].findAll('div', class_='gl-el'):
        #print(sent.findAll(class_='gl-data-tx-2')[0])
        lex_corpus.extend([x.text for x in sent.findAll(class_='gl-data-tx-2')[0].findAll('td')])
        gram_corpus.extend([x.text for x in sent.findAll(class_='gl-data-tx-3')[0].findAll('td')])
#//*[@id="node-text-full-group-gloss-wrap"]
#//*[@id="node-text-full-group-gloss"]/div
#class="group-gloss-wrap field-group-html-element"
#class="gl-data-tx-2"
##node-text-full-group-text > div > div > div > div > p
#class="collapsible group-text field-group-div speed-slow effect-none fieldgroup-effects-processed collapse-close"
#class="required-fields text-wrap"
```

```python
from collections import Counter
Counter(gram_corpus).most_common()
```
