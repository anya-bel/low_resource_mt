# Russian Word2Vec training

```python
common_texts = [word_tokenize(text) for text in sentences]
```

```python
model = Word2Vec(sentences=common_texts, vector_size=100, window=5, min_count=3, workers=-1)
model.save("russian_word2vec.model")
```
